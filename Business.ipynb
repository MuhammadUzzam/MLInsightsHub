<h3 style="background-color:#f5f5f5;font-family:newtimeroman;color:#0066cc;text-align:center">
Sales Analysis by Category, Sub-Category, Region, and Ship Mode using bar chart.</h3>

```python
# Create the figure and subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))  
# Plot 1: Sales by Sub-Category
# Group the data by Sub-Category and calculate the sum of Sales
sub_category_data = store.groupby('Sub-Category').sum()
# Sort the data by Sales in descending order
sub_category_data = sub_category_data.sort_values('Sales', ascending=False)
# Define the colormap with increased saturation
colormap = plt.cm.get_cmap('viridis')
colormap = colormap(np.linspace(0, 1, len(sub_category_data)))
colormap[:, 1] = np.clip(colormap[:, 1] + 0.2, 0, 1)  # Increase the saturation 
bars1 = ax1.bar(sub_category_data.index, sub_category_data['Sales'], color=colormap)
ax1.set_title('Sales by Sub-Category', fontsize=16, fontweight='bold')
ax1.set_xlabel('Sub-Category', fontsize=12)
ax1.set_ylabel('Sales', fontsize=12)
ax1.set_xticklabels(sub_category_data.index, rotation=70)
ax1.set_facecolor('#FFCC99')
# Plot 2: Sales by Region
store_sorted = store.sort_values(by='Sales', ascending=False)
bars2 = ax2.bar(store_sorted['Region'], store_sorted['Sales'], color='#05eff7')
ax2.set_title('Sales by Region', fontsize=16, fontweight='bold')
ax2.set_xlabel('Region', fontsize=12)
ax2.set_ylabel('Sales', fontsize=12)
ax2.set_facecolor('#dfedd8')
# Plot 3: Sales by Category
bars3 = ax3.bar(store_sorted['Category'], store_sorted['Sales'], color='#f5baba')
ax3.set_title('Sales by Category', fontsize=16, fontweight='bold')
ax3.set_xlabel('Category', fontsize=12)
ax3.set_ylabel('Sales', fontsize=12)
ax3.set_facecolor('#F2F2F2')
# Plot 4: Sales by Ship Mode
bars4 = ax4.bar(store_sorted['Ship Mode'], store_sorted['Sales'], color='#2ecc71')
ax4.set_title('Sales by Ship Mode', fontsize=16, fontweight='bold')  # Fixed typo in the title
ax4.set_xlabel('Ship Mode', fontsize=12)
ax4.set_ylabel('Sales', fontsize=12)
ax4.set_facecolor('#ecf0f1')
# Adjust the layout to prevent overlapping of the subplots
fig.tight_layout()
# Display the plot
plt.show()
```

<h3 style="background-color:#f5f5f5;font-family:newtimeroman;color:#0066cc;text-align:center">
Profit Analysis by Category, Sub-Category, Region, and Ship Mode using line chart and bar chart.</h3>

```python
# Calculate profit by Category
category_profit = store.groupby('Category')['Profit'].sum()
# Calculate profit by Sub-Category
sub_category_profit = store.groupby('Sub-Category')['Profit'].sum()
# Calculate profit by Region
region_profit = store.groupby('Region')['Profit'].sum()
# Calculate profit by Ship Mode
ship_mode_profit = store.groupby('Ship Mode')['Profit'].sum()
# Create the figure and subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 10), facecolor='lightgrey')
# Plot 1: Profit by Category
ax1.plot(category_profit.index, category_profit, marker='o', linewidth=2, markersize=8, color='blue')
ax1.fill_between(category_profit.index, category_profit, color='lightblue')
ax1.set_title('Profit by Category', fontsize=16, fontweight='bold')
ax1.set_xlabel('Category', fontsize=12)
ax1.set_ylabel('Profit', fontsize=12)
ax1.set_xticklabels(category_profit.index)
ax1.grid(True)
# Plot 2: Profit by Sub-Category
ax2.plot(sub_category_profit.index, sub_category_profit, marker='o', linewidth=2, markersize=8, color='green')
ax2.fill_between(sub_category_profit.index, sub_category_profit, color='lightgreen')
ax2.set_title('Profit by Sub-Category', fontsize=16, fontweight='bold')
ax2.set_xlabel('Sub-Category', fontsize=12)
ax2.set_ylabel('Profit', fontsize=12)
ax2.set_xticklabels(sub_category_profit.index, rotation=45)
ax2.grid(True)
# Plot 3: Profit by Region
ax3.plot(region_profit.index, region_profit, marker='o', linewidth=2, markersize=8, color='purple')
ax3.fill_between(region_profit.index, region_profit, color='#A569BD')
ax3.set_title('Profit by Region', fontsize=16, fontweight='bold')
ax3.set_xlabel('Region', fontsize=12)
ax3.set_ylabel('Profit', fontsize=12)
ax3.set_xticklabels(region_profit.index)
ax3.grid(True)
# Plot 4: Profit by Ship Mode
ax4.plot(ship_mode_profit.index, ship_mode_profit, marker='o', linewidth=2, markersize=8, color='#FF8C00')
ax4.fill_between(ship_mode_profit.index, ship_mode_profit, color='#FFCC99')
ax4.set_title('Profit by Ship Mode', fontsize=16, fontweight='bold')
ax4.set_xlabel('Ship Mode', fontsize=12)
ax4.set_ylabel('Profit', fontsize=12)
ax4.set_xticklabels(ship_mode_profit.index)
ax4.grid(True)
# Adjust the layout to prevent overlapping of the subplots
fig.tight_layout()
# Display the plot
plt.show()
```
